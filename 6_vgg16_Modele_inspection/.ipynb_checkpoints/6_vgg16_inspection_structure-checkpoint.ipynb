# Importation des bibliothèques 

```python
import keras
from keras.datasets import cifar10
from keras.utils import to_categorical
from keras_preprocessing import image
from keras.applications import vgg16
import joblib
from keras.models import Model
from matplotlib import pyplot
from numpy import expand_dims
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from sklearn.model_selection import train_test_split
```

# Chargement du modèle CNN VGG16

```python
# charger le modèle
features_extraction_model = vgg16.VGG16(weights='imagenet', include_top=False,input_shape=(32,32,3))
# résumer le modèle
features_extraction_model.summary()
```

# Visualisation des filtres

```python
# résumer les filtres dans chaque couche convolutive
# résumer les formes des filtres
for layer in features_extraction_model.layers:
	# vérifier la couche convolutive
	if 'conv' not in layer.name:
		continue
	# obtenir des poids de filtre
	filters, biases = layer.get_weights()
	print(layer.name, filters.shape)
```

```python
# récupérer des poids dans la deuxième couche cachée
filters, biases = features_extraction_model.layers[1].get_weights()
# normaliser les valeurs des filtres à 0-1 afin de pouvoir les visualiser
f_min, f_max = filters.min(), filters.max()
filters = (filters - f_min) / (f_max - f_min)
# afficher les premiers filtres
n_filters, ix = 10, 1
for i in range(n_filters):
	# obtenir le filtre
	f = filters[:, :, :, i]
	# afficher chaque canal séparément
	for j in range(3):
		# préciser la sous-parcelle et la rotation de l'axe
		ax = pyplot.subplot(n_filters, 3, ix)
		ax.set_xticks([])
		ax.set_yticks([])
		# tracer le canal du filtre en échelle de gris
		pyplot.imshow(f[:, :, j], cmap='gray')
		ix += 1
# afficher la figure
pyplot.show()
```

# Visualiser des caractéristiques d'une image en entrée dans chaque bloc de convolution

```python
# visualiser les cartes de caractéristiques produites par chaque bloc dans le modèle vgg avec 224x224

# charger le modèle
features_extraction_model_blocs = vgg16.VGG16(weights='imagenet', include_top=False,input_shape=(32,32,3))
# redéfinir le modèle en sortie juste après la première couche cachée
ixs = [2, 5, 9, 13, 17]
outputs = [features_extraction_model_blocs.layers[i].output for i in ixs]
blocs_features_model  = Model(inputs=features_extraction_model_blocs.inputs, outputs=outputs)
# charger l'image avec la forme requise
img = load_img("my_images/c.jpg", target_size=(224, 224))
# convertir l'image en un tableau
img = img_to_array(img)
# élargir les dimensions de manière à ce qu'elles représentent un "échantillon" unique
img = expand_dims(img, axis=0)
# préparer l'image (par exemple, mettre à l'échelle les valeurs des pixels pour le vgg)
img = preprocess_input(img)
# obtenir la carte des caractéristiques de la première couche cachée
blocs_feature_maps = blocs_features_model .predict(img)
# tracer le résultat de chaque bloc
square = 8
for fmap in blocs_feature_maps:
	# tracer les 64 cartes dans un carré de 8x8
	ix = 1
	for _ in range(square):
		for _ in range(square):
			# préciser la sous-parcelle et la rotation de l'axe
			ax = pyplot.subplot(square, square, ix)
			ax.set_xticks([])
			ax.set_yticks([])
			# tracer le canal du filtre en échelle de gris
			pyplot.imshow(fmap[0, :, :, ix-1], cmap='gray')
			ix += 1
	# afficher la figure
	pyplot.show()
```
